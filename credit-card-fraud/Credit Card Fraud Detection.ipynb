```python
# importing required libraries
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

%matplotlib inline
```

```python
# loading dataset
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df.describe()
```

Oberservation 
<br>
1. All the columns are normalised except Time and Amount.

```python
# scaling the unscaled columns 
rbs = RobustScaler()  # Instantiating Robust Scaler object 

df_small = df[['Time','Amount']]
df_small = pd.DataFrame(rbs.fit_transform(df_small))  # transforming data using robust scaler 
# alternate way is to use StandardScaler, but RobustScaler is lesser prone to outliers outliers 

df_small.columns = ['scaled_time','scaled_amount']
df = pd.concat([df,df_small],axis=1)

df.drop(['Time','Amount'],axis=1,inplace=True)

df.head()
```

```python
# dataset visualisation => the dataset as of yet is unbalanced 
sns.countplot(df['Class'], palette = 'Set3')
plt.show()
```

```python
df['Class'].value_counts()
```

<br>
Class 1 - Fradulent transactions
<br>
Class 2 - Non-fradulent transactions
<br><br>
This dataset is highly unbalanced because we have 284315 non-fraudulent transactions and 492 fraudulent transactions. 
<br>
If we train the model on this data, then it will predict the non-fraudulent transactions every time, and the overall accuracy will be very high. But, it would be a very bad model. Thus, the dataset has to be balanced first.  

```python
# Balancing the Dataset 
non_fraud = df[df['Class'] == 0]
fraud = df[df['Class'] == 1]

non_fraud = non_fraud.sample(frac = 1)

non_fraud = non_fraud[:492]

new_df = pd.concat([non_fraud, fraud])
new_df = new_df.sample(frac = 1)
```

df.sample randomly takes out some fraction of data from the dataframe.
<br>
frac = 1 -> 100% of the data
<br>
This step helps in the shuffling of data. 
<br><br>
After shuffling, we only take the first 492 instances of non_fraud because there are only 492 instances of fraudulent transactions. This balances the dataset. 
<br>
We shuffle the data after joining the two datasets. 

```python
# visualising the balanced dataset
sns.countplot(new_df['Class'], palette = 'Set3')
plt.show()
```

```python
new_df['Class'].value_counts()  # equal instances of 0s and 1s
```

```python
# Plotting Amount v. Class 
sns.jointplot(new_df['scaled_amount'], new_df['Class'])
plt.show()
```

All the 1s are on the left-hand-side (LHS) of the plot, which means a lesser amount. 
<br>
Thus, most of the fraudulent transactions were done for a lesser amount. 

```python
# train_test_split in the ratio of 80% to 20%
X = new_df.drop('Class', axis = 1)
y = new_df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
# training the Logistic Regression model to detect Credit Card fraud
from sklearn.linear_model import LogisticRegression 

lr = LogisticRegression()
lr.fit(X_train, y_train)

pred = lr.predict(X_test)

print(classification_report(y_test, pred))
print('\n\n')
print(confusion_matrix(y_test, pred))
print('\n')
print('accuracy is --> ', round(accuracy_score(y_test, pred) * 100, 2))
```

<br>
In this case, Logistic Regression is better than Random Forest or Decision Tree classifiers.
